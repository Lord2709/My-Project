```python
import seaborn as sns
x = 29700313
y = 28491670
z= 381931
active= x-y
print('Total Confirmed cases =',x)
print('Total Cured cases =',y)
print('Total Active cases =',active)
print('Total Death cases =',z)
barp = sns.barplot(x=['Confirmed','Cured','Deaths','active'],y=[x,y,z,active])
barp.set_yticklabels(labels=(barp.get_yticks()*1).astype(int))
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('/content/Latest Covid-19 India Status.csv')
df
```

```python
df1 = df[['State/UTs','Total Cases','Deaths']][:10]
df1
```

```python
from matplotlib import pyplot
a4_dims = (15, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
barp = sns.barplot(data=df1,x='State/UTs',y='Total Cases',ax=ax)
barp.set_yticklabels(labels=(barp.get_yticks()*1).astype(int))
barp = sns.set_theme(style="whitegrid")
ax.set_title("Top 10 Contaminated states")
plt.show()
```

```python
df2 = df[['State/UTs','Total Cases','Deaths','Discharged']][:10]
df2
```

```python
df2['Fatality-Ratio'] = df2['Deaths']/df2['Total Cases']
a4_dims = (17, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
plt.title('Fatality-Ratio of Top 10 Contaminated States',size=15)
sns.pointplot(data=df2,x='State/UTs',y='Fatality-Ratio',ax=ax,color='Red')
plt.show()
```

```python
df2['Cure-Ratio'] = df2['Discharged']/df2['Total Cases']
a4_dims = (17, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
plt.title('Cure-Ratio of Top 10 Contaminated States',size=15)
sns.pointplot(data=df2,x='State/UTs',y='Cure-Ratio',ax=ax,color='Green')
plt.show()
```
